# Tratamiento de Datos de Películas y Valoraciones de Usuarios

Utilizaremos Pandas para manejar y transformar los datos

```python
import pandas as pd
```

## Vamos a trabajar con un conjunto de datos que contiene información sobre películas y valoraciones de usuarios.

El fichero *movies.csv* contiene las siguientes columnas:
* **movieId**: identificador de la película.
* **title**: título de la película.
* **genres**: géneros de la película. Están separados por el carácter "|".

El fichero *ratings.csv* contiene las siguientes columnas:

* **userId**: identificador del usuario que ha valorado la película.
* **movieId**: identificador de la película.
* **rating**: valoración de la película por el usuario, en una escala de 0.5 a 5.
* **timestamp**: fecha y hora en la que se realizó la valoración. Está en formato UNIX, es decir, el número de segundos que han pasado desde el 1 de enero de 1970. Luego veremos cómo convertirlo a un formato más legible.

### 1. Carga de datos

Leemos los ficheros de datos y los almacenamos en dos DataFrames de Pandas. os llamamoss `movies` y `ratings`

```python
# Leemos los ficheros de películas y valoraciones almacenándolos en dos DataFrames:

movies = pd.read_csv("https://raw.githubusercontent.com/Luz-Ensenat/movie-ratings-data-analysis/main/data/movies.csv")

ratings = pd.read_csv("https://raw.githubusercontent.com/Luz-Ensenat/movie-ratings-data-analysis/main/data/ratings.csv")

# Vemos los tamaños que tiene cada uno para saber el nº de películas y de valoraciones:

print (f"Nº de películas en el DataFrame movies: {movies.shape[0]}")
print (f"Nº de Valoraciones en el DataFrame ratings: {ratings.shape[0]}")

# Vemos las columnas que contine cada DataFrame:

print (f"Las columnas del DataFrame movies son: {movies.columns}")
print (f"Las columnas del DataFrame ratings son: {ratings.columns}")
```

### 2. Union de DataFrames

Añadimos el título y los géneros de las películas a la tabla de valoraciones de tal forma que tengamos una única tabla con toda la información. La llamaremos `movie_ratings`. Las películas que no tengan valoraciones o ratings de películas que no conocemos no nos interesan.

```python
# Unimos los dos DataFrames por movieId, la columna común para hacer la unión y teniendo en cuenta con inner sólo las que coincidan en ambos:

movie_ratings = ratings.merge(movies[['movieId', 'title', 'genres']], on='movieId', how='left')

# Vemos cómo ha quedado nuestro DataFrame final:

movie_ratings
```

### 3. Ejemoplo: cálculo de las valoraciones que ha recibido la película "Dark Knight, The (2008)" y cuántas de ellas son de 5 estrellas

```python
# Filtramos las valoraciones de la película "Dark Knight, The (2008)" e imprimimos el DataFrame resultante:

dk_ratings = movie_ratings[movie_ratings['title'] == 'Dark Knight, The (2008)']

dk_ratings
```

```python
# Medimos el número total de valoraciones y lo imprimimos:

total_valoraciones_dk = dk_ratings.shape[0]

print("Número total de valoraciones de la película Dark Knight, The (2008):", total_valoraciones_dk)
```

```python
# Realizamos un doble filtro sobre la película "Dark Knight, The (2008)", dejando sólo las valoraciones de 5 estrellas e imprimimos el DataFrame resultante:

dk_ratings_5 = movie_ratings[(movie_ratings['title'] == 'Dark Knight, The (2008)') & (movie_ratings['rating'] == 5.0)]

dk_ratings_5
```

```python
# Medimos el número total de valoraciones 5 estrellas y lo imprimimos:

valoraciones_5_dk = dk_ratings_5.shape[0]

print("Total de valoraciones 5 estrellas de la película Dark Knight, The (2008):", valoraciones_5_dk)
```

### 4. Cálculo de la valoracion media y el número de valoraciones de cada película

Almacenamos el resultado en un nuevo DataFrame llamado `movie_stats`.

```python
# Creamos el DataFrame movie_stats grupando las películas por título mediante groupby y calculando la media y el número de valoraciones mediante mean y size:

movie_stats = movie_ratings.groupby('title').agg(rating_count = ('rating', 'size'), rating_mean = ('rating', 'mean')).reset_index()

movie_stats
```

### 5. Cálculo de la película que tiene la valoración media más alta

Para tener un resutltado más fiable filtramos aquellas películas que tengan al menos 100 valoraciones y extraemos el top 10 de películas con la valoración media más alta.

```python
# Filtramos películas con al menos 100 valoraciones:

filtered_stats = movie_stats[movie_stats['rating_count'] >= 100]

# Ordenamos por valoración media descendente mediante sort y extraemos el top 10 mediante head:

top_10_peliculas = filtered_stats.sort_values(by='rating_mean', ascending=False).head(10)

top_10_peliculas
```

### 6. Extracción del año de la película

El año de la película se encuentra entre paréntesis en el título. Añadiremos una columna al DataFrame `movies` con el año de la película asumiendo que siempre se encuentra en la misma posición. Utilizaremos una función lambda para extraer el año del título. Una vez que lo hayamos extraído, eliminaremos el año del título.

```python
# Importamos el módulo re para buscar el año dentro del título de las películas:

import re
import numpy as np

# Extraemos el año entre paréntesis si existe (porque si no existe nos da error) y lo guardamos en una nueva columna 'year':

def extrae_año(título):
    m = re.search(r' (\d4)', título)
    return m.group(1) if m else np.nan

movies['year'] = movies['title'].apply(extrae_año)

movies
```

```python
# Eliminamos el año y los paréntesis del título mediante una función lambda:

movies['title'] = movies['title'].apply(lambda x: re.sub(r'\s*\(\d{4}\)', '', x))

movies
```

### 7. Filtro por género

Filtraremos las películas de un género concreto. Por ejemplo, obtendremos las películas de acción de 1993.

```python
# Primero convertimos a número entero la columna "year" y, si no parece, lo sustituimos por NaN (Not a Number) mediante coerce. Usamos loc para evitar el error SettingWithCopyWarning:

movies.loc[:, 'year'] = pd.to_numeric(movies['year'], errors='coerce')

# Filtramos las películas de acción (genres) de 1993 (year) mediante  el método str.contains e imprimimos:

action_93 = movies[movies['genres'].str.contains('Action') & (movies['year']==1993)]

action_93
```

### 8. Tratamiento columna de fecha UNIX

El campo `timestamp` de la tabla `ratings` está en formato UNIX. Crearemos una nueva columna en la tabla `ratings` llamada `date` que contenga la fecha en un formato legible. Calcularemos cuándo fue la primera y la última valoración.

```python
# Creamos una nueva columna "date" legible convirtiéndolo mediante to_datetime de Pandas:

ratings['date'] = pd.to_datetime(ratings['timestamp'], unit='s')

ratings
```

```python
# Imprimimos cuándo fué la primera y la última valoraciones:

print("Primera valoración:", ratings['date'].min())
print("Última valoración:",    ratings['date'].max())
```
