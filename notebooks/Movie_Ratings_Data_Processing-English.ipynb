# Movie and User Ratings Data Processing

We will use Pandas to manage and transform the data.

```python
import pandas as pd
```

## We will work with a dataset containing information about movies and user ratings.

The file *movies.csv* contains the following columns:

* **movieId**: movie identifier.
* **title**: movie title.
* **genres**: movie genres. They are separated by the "|" character.

The file *ratings.csv* contains the following columns:

* **userId**: identifier of the user who rated the movie.
* **movieId**: movie identifier.
* **rating**: movie rating given by the user, on a scale from 0.5 to 5.
* **timestamp**: date and time when the rating was made. It is stored in UNIX format, meaning the number of seconds elapsed since January 1, 1970. Later, we will see how to convert it into a more readable format.


### 1. Data Loading

We read the data files and store them in two Pandas DataFrames called `movies` and `ratings`.


```python
# We read the movie and ratings files and store them in two DataFrames:

movies = pd.read_csv("https://raw.githubusercontent.com/Luz-Ensenat/movie-ratings-data-analysis/main/data/movies.csv")

ratings = pd.read_csv("https://raw.githubusercontent.com/Luz-Ensenat/movie-ratings-data-analysis/main/data/ratings.csv")

# We inspect the size of each DataFrame to know the number of movies and ratings:

print(f"Number of movies in the movies DataFrame: {movies.shape[0]}")
print(f"Number of ratings in the ratings DataFrame: {ratings.shape[0]}")

# We inspect the columns contained in each DataFrame:

print(f"The columns in the movies DataFrame are: {movies.columns}")
print(f"The columns in the ratings DataFrame are: {ratings.columns}")
```

### 2. DataFrame Merge

We add the movie titles and genres to the ratings table so that we have a single table containing all the information. We will call it `movie_ratings`.

Movies without ratings or ratings for unknown movies are not relevant for our analysis.


```python
# We merge the two DataFrames using movieId as the common column and apply a left join:

movie_ratings = ratings.merge(movies[['movieId', 'title', 'genres']], on='movieId', how='left')

# We inspect the resulting final DataFrame:

movie_ratings
```

### 3. Example: calculation of the ratings received by the movie "Dark Knight, The (2008)" and how many of them are 5-star ratings


```python
# We filter the ratings for the movie "Dark Knight, The (2008)" and display the resulting DataFrame:

dk_ratings = movie_ratings[movie_ratings['title'] == 'Dark Knight, The (2008)']

dk_ratings
```

```python
# We calculate the total number of ratings and print the result:

total_ratings_dk = dk_ratings.shape[0]

print("Total number of ratings for the movie Dark Knight, The (2008):", total_ratings_dk)
```

```python
# We apply a double filter to the movie "Dark Knight, The (2008)", keeping only 5-star ratings, and display the resulting DataFrame:

dk_ratings_5 = movie_ratings[(movie_ratings['title'] == 'Dark Knight, The (2008)') & (movie_ratings['rating'] == 5.0)]

dk_ratings_5
```

```python
# We calculate the total number of 5-star ratings and print the result:

ratings_5_dk = dk_ratings_5.shape[0]

print("Total number of 5-star ratings for the movie Dark Knight, The (2008):", ratings_5_dk)
```

### 4. Calculation of the average rating and the number of ratings for each movie

We store the result in a new DataFrame called `movie_stats`.


```python
# We create the movie_stats DataFrame by grouping movies by title using groupby and calculating the average rating and number of ratings using mean and size:

movie_stats = movie_ratings.groupby('title').agg(
    rating_count=('rating', 'size'),
    rating_mean=('rating', 'mean')
).reset_index()

movie_stats
```

### 5. Calculation of the movie with the highest average rating

To obtain a more reliable result, we filter movies with at least 100 ratings and extract the top 10 movies with the highest average rating.


```python
# We filter movies with at least 100 ratings:

filtered_stats = movie_stats[movie_stats['rating_count'] >= 100]

# We sort by average rating in descending order using sort_values and extract the top 10 using head:

top_10_movies = filtered_stats.sort_values(by='rating_mean', ascending=False).head(10)

top_10_movies
```

### 6. Extraction of the movie release year

The movie release year appears in parentheses in the title. We will add a column to the `movies` DataFrame containing the movie year, assuming it is always located in the same position.

We will use a lambda function to extract the year from the title. Once extracted, we will remove the year from the title.


```python
# We import the re module to search for the year within movie titles:

import re
import numpy as np

# We extract the year in parentheses if it exists (otherwise it would raise an error) and store it in a new column called 'year':

def extract_year(title):
    m = re.search(r'\((\d{4})\)', title)
    return m.group(1) if m else np.nan

movies['year'] = movies['title'].apply(extract_year)

movies
```

```python
# We remove the year and parentheses from the title using a lambda function:

movies['title'] = movies['title'].apply(lambda x: re.sub(r'\s*\(\d{4}\)', '', x))

movies
```

### 7. Genre Filtering

We will filter movies belonging to a specific genre. For example, we will obtain action movies released in 1993.


```python
# First, we convert the "year" column to integer format. If conversion fails, the value is replaced with NaN (Not a Number) using coerce.
# We use loc to avoid the SettingWithCopyWarning error:

movies.loc[:, 'year'] = pd.to_numeric(movies['year'], errors='coerce')

# We filter Action movies released in 1993 using str.contains and display the result:

action_93 = movies[movies['genres'].str.contains('Action') & (movies['year'] == 1993)]

action_93
```

### 8. UNIX Date Column Processing

The `timestamp` field in the `ratings` table is stored in UNIX format.

We will create a new column in the `ratings` DataFrame called `date` containing the date in a readable format.

We will also calculate when the first and last ratings were made.


```python
# We create a new readable "date" column by converting the UNIX timestamp using Pandas to_datetime:

ratings['date'] = pd.to_datetime(ratings['timestamp'], unit='s')

ratings
```

```python
# We print the dates of the first and last ratings:

print("First rating:", ratings['date'].min())
print("Last rating:", ratings['date'].max())
```
